# Polynomial regression

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model as sk_lm
from sklearn import model_selection as sk_msel
from sklearn import metrics as sk_metrics
from sklearn import preprocessing as sk_pre
import statsmodels.api as sm
```

```python
df_wage = pd.read_csv('data/Wage.csv')
df_wage.drop(columns='Unnamed: 0', inplace=True)
df_wage.head()
```

```python
# Polynomial transformation of age predictor
X1=sk_pre.PolynomialFeatures(1).fit_transform(df_wage[['age']])
X2=sk_pre.PolynomialFeatures(2).fit_transform(df_wage[['age']])
X3=sk_pre.PolynomialFeatures(3).fit_transform(df_wage[['age']])
X4=sk_pre.PolynomialFeatures(4).fit_transform(df_wage[['age']])
X5=sk_pre.PolynomialFeatures(5).fit_transform(df_wage[['age']])
```

```python
y=df_wage.wage
```

```python
polyfit1=sm.OLS(y,X1).fit()
polyfit2=sm.OLS(y,X2).fit()
polyfit3=sm.OLS(y,X3).fit()
polyfit4=sm.OLS(y,X4).fit()
polyfit5=sm.OLS(y,X5).fit()
```

```python
polyfit3.summary()
```

```python
sm.stats.anova_lm(polyfit1, polyfit2, polyfit3, polyfit4, polyfit5, typ=1)
```
