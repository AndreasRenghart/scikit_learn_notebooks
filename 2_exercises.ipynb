```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

## 1. College Dataset

```python
df_c=pd.read_csv('data/College.csv')
```

```python
df_c.head()
```

```python
df_c.rename(columns={'Unnamed: 0': 'College'}, inplace=True)
```

```python
df_c.head()
```

```python
df_c.describe()
```

### Scatter Matrix

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df_c[['Apps', 'Room.Board', 'Expend', 'Books', 'S.F.Ratio']],
               alpha=0.8, figsize=(10, 10), diagonal='kde')
```

### Side-by-side Boxplot

```python
sns.boxplot(x='Private', y='Outstate', data=df_c)
```

### Binning Data

```python
bins=[0, 50, np.inf]
labels=['No', 'Yes']
df_c['Elite']=pd.cut(df_c['Top25perc'], bins=bins, labels=labels)
df_c.head()
```

```python
df_c.groupby('Elite').size()
```

```python
sns.boxplot(x='Elite', y='Outstate', data=df_c)
```

### Histograms

```python
# With pandas
df_c['Room.Board'].plot(kind='hist', bins=20)
```

```python
# With seaborn
sns.distplot(df_c['Room.Board'], bins=20, kde=False)
```

```python
sns.distplot(df_c['Room.Board'], bins=20)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, sharey=True, figsize=(12,5))
sns.distplot(df_c['Enroll'], kde=False, bins=20, ax=ax1)
sns.distplot(df_c['Expend'], kde=False, bins=20, ax=ax2)
sns.distplot(df_c['Room.Board'], kde=False, bins=20, ax=ax3)
```
