# What is Statistical Learning?

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df_ad=pd.read_csv('data/Advertising.csv')
```

```python
df_ad.head()
```

**Sales in relation to TV, Radio and Newspaper advertising budgets**

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, sharey=True, figsize=(12,5))
sns.regplot(x='TV', y='Sales', data=df_ad, ax=ax1)
sns.regplot(x='Radio', y='Sales', data=df_ad, ax=ax2)
sns.regplot(x='Newspaper', y='Sales', data=df_ad, ax=ax3)
```
